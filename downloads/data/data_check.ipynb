```python
import glob
import json
import pprint
```

```python
data_path="./aihub/admin_mrc/**/*.json"
```

```python
kor_list = glob.glob(data_path)
print(kor_list)
```

```python
docs = []
for doc in kor_list:
    if 'span' not in doc:
        # normal korquad style
        continue
    with open( doc, "r",encoding='utf8') as fr:
        docs.append(json.load(fr))
```

```python
# keys of a document.``
docs[0]['data'][0]
```

```python
print(f"len of train data {len(docs[1]['data'])}")
print(f"len of dev data {len(docs[0]['data'])}")
```

```python
docs[1]['data'][0]['paragraphs'][0]['qas']
```

```python
prag_list = []
check_ctx = []
check_docs = []

for doc in docs[0]['data']:
    form = {
    "title":"",
    "len_pragraphs":0,
    "len_ctxs":[],
    "len_qas":[]
    }
    form["title"] = doc['title']
    form['len_pragraphs'] = len(doc['paragraphs'])
    for prag in doc['paragraphs']:
        form["len_ctxs"].append(len(prag['context'].split(" ")))
        check_ctx.append(form["len_ctxs"][-1])
        check_docs.append((doc["title"], prag["context"]))
        form["len_qas"].append(len(prag["qas"]))
    prag_list.append(form)
```

```python
tmp_idx = check_ctx.index(max(check_ctx))
print(max(check_ctx), min(check_ctx))
check_docs[tmp_idx]
```
